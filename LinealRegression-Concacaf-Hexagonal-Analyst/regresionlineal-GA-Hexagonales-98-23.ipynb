```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('hexagonal-98-22.csv')
```

```python
data
```

```python
y = data.Rk
x = data.GA.values.reshape(-1,1)

print(y.shape, x.shape)
```

```python
model = LinearRegression().fit(x,y)
r_sq = model.score(x,y)
print(r_sq)
```

```python
# y = mx + b
b = model.intercept_
m = model.coef_
y_pred = m*x +b
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
plt.grid(c='black')
plt.plot(x,y_pred,'red')
plt.ylabel('Rank')
plt.xlabel('Goles en contra')
plt.title('Evaluacion de relacion entre goles en contra y puntos hexagonales 98 hasta 2022')
plt.show()
```

```python
#y = m * x + b
for i in range(0, 25):
    yCustom = m * (i) + b
    print(i,"",yCustom)
```
